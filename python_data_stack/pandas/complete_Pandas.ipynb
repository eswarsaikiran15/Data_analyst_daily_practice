# dropping rows with NaN values

```python
df.shape
```

```python
df2=df.dropna() # default axis =0
```

dropna() is the opposite kind of operation compared to isnull().

Instead of detecting missing values, dropna() removes them.

```python
df2.shape
```

# droppping columns with NaN values

```python
df3=df.dropna(axis=1)  #axis =1 for columns
```

```python
df3.shape
```

```python
df.dropna(how='any')  # if any row value is null then remove that row
```

```python
df.dropna(how='all') # if all rows values is null then remove that row
```

```python
df.dropna(inplace=True) #Remove rows with missing values (NaN) and directly modify the original DataFrame
```

```python
df.shape
```

# Filling the null values


```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Diabetes Missing Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Glucose':0,'Diastolic_BP':24,'Skin_Fold':34})
```

```python
df.fillna(method='ffill') #forward fill -previous value is filled inplace of null value
```

```python
df.fillna(method='ffill',axis=1) #forward fill -if we put axis =1 then previous column same row value is filled in cureent column
```

```python
df['Glucose'].fillna(value=df['Diabetes_Pedigree'].mean())
```

# Povit table

```python
import pandas as pd
df=pd.read_csv('/content/Diabetes Missing Data.csv')
df.head()
```

```python
pd.pivot_table(df,index='Glucose')  #it internally behaves like df.groupby('Glucose').mean()
```

```python
pd.pivot_table(df,index='Glucose',aggfunc='sum') #Sums all other numeric columns for each group
```

```python
pd.pivot_table(df,index='Glucose',columns='Pregnant')
```
